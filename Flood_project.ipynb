```python
#import the data from the directory
import pandas as pd
df = pd.read_excel("FloodArchive.xlsx")
df.head()
```

```python
# checking fr data info
df.info()
```

```python
#replacing 0 with None in the GlideNumber column and same for other columns if necessary
df["GlideNumber"] = df["GlideNumber"].replace( 0, None)
```

```python
df["OtherCountry"] = df["OtherCountry"].replace( 0, None)
```

```python
#checking for duplicates
df.duplicated().sum()
```

```python
#renaming the columns for better understanding
df.rename(columns={
    'lat': 'Latitude',
    'long': 'Longitude' 
})
```

```python
# checking for missing values
df[df['Validation'] == 0].groupby('GlideNumber').size()
```

```python
# filling the missing values in the Validation column with "GLIDE/OCHA"
df.loc[
    (df["Validation"].isna()) | (df["Validation"] == 0),
    "Validation"
] = "GLIDE/OCHA"
```

```python
df[df['Validation'] == 0].groupby('GlideNumber').size()
```

```python
# exporting the cleaned data to a new CSV file
df.to_csv('flood_data_cleaned.csv', index=False)
```
